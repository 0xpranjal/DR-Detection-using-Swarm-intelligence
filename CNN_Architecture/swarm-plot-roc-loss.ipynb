```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import json
```

```python
def read_data(path):
    f = open(path)
    data = json.load(f,)
    num_epochs = len(data['train_loss'])
    epochs = [i+1 for i in range(num_epochs)]
    data['epochs'] = epochs
    return data
```

```python
eff_b5 = read_data('../input/swarm-plots/efficientnet_b5_ROC_LOSS.json')
eff_b5['name'] = 'Efficientnet B5'
res_200 = read_data('../input/swarm-plots/resnet200d_ROC_LOSS.json')
res_200['name'] = 'Resnet 200D'
data_list = [eff_b5,res_200]
```

```python
def plot_data(data_list,param):
    fig = go.Figure()
    for i in range(len(data_list)):
        temp = data_list[i]
        fig.add_trace(go.Scatter(
                            x=temp['epochs'][:10], y=temp[param][:10],
                            mode='lines+markers',
                            name=temp['name']
                        )
            )
    if param == 'val_roc':   
        fig.update_layout(title="Validation AUC ROC Across Epochs",
                   xaxis_title='Epoch',
                   yaxis_title='Validation AUC ROC')
    elif param == 'train_loss':
        fig.update_layout(title="Train Loss Across Epochs",
                   xaxis_title='Epoch',
                   yaxis_title='Train Loss')
    elif param == 'val_loss':
        fig.update_layout(title="Valid Loss Across Epochs",
                   xaxis_title='Epoch',
                   yaxis_title='Validation Loss')

        
    fig.show()
```

```python
plot_data(data_list,'train_loss')
```

```python
plot_data(data_list,'train_loss')
```

```python
plot_data(data_list,'val_loss')
```
